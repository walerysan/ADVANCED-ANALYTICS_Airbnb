# 2 Exploring relationships

Create a correlation matrix using pandas

```python
df_airbnb.corr()
```

Create a correlation heatmap using seaborn

```python
# Create subset exluding columns

sub = df_airbnb[['price', 'distance_to_beach','minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'number_of_reviews_ltm']]
```

```python
sub
```

```python
# In the first step, create a subplot with matplotlib

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

Comment: 
Purple equates to 0, or no relationship between the variables, dark purple and black equate to negative correlations, and orange and light orange equate to positive correlations.

The coefficient of 0.69 between number_of_reviews and number_of_reviews_ltm (in the last 12 months) indicates medium to strong relation: the higher the number of reviews per listing, the higher the number of those reviews was left by guests in the last 12 months. This relation can help to indicate active listings.

There is a very weak positive relationship of 0.014 between the price per night and number of properties managed by one host: the price goes slightly down for multiple listings hosted by one owner, which can implicate that owners with only one listing put higher price tags per night.

The coefficient of -0.093 between price and number of reviews in the last 12 months indicates a weak negative relation which means that the higher price of the property most likely leads to less bookings thus lower number of left reviews. 

A weak negative relation of -0.12 between minimum nights and number of reviews in the last 12 months

# Categorical Plots

Creating a categorical variable that splits the "price" column into categories

```python
# Use a histogram to visualize the distribution of the variables

sns.distplot(df_airbnb['price'], bins = 250)
```

```python
# Due to small count of extremely values of prices, the histogram doesn't show me the potential split, so I am using the frequency count to determine the categories
```

```python
df_airbnb['price'].value_counts(dropna = False)
```

```python
df_airbnb.loc[df_airbnb['price'] < 80, 'price_category'] = 'low price'
```

```python
df_airbnb.loc[(df_airbnb['price'] >= 80) & (df_airbnb['price'] <= 150), 'price_category'] = 'middle price'
```

```python
df_airbnb.loc[(df_airbnb['price'] > 150) & (df_airbnb['price'] <= 500), 'price_category'] = 'high price'
```

```python
df_airbnb.loc[df_airbnb['price'] >500, 'price_category'] = 'extreme price'
```

```python
df_airbnb['price_category'].value_counts(dropna = False)
```

```python
df_airbnb.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="number_of_reviews", y="price", hue="price_category", data=df_airbnb)
```

```python
sns.set(style="ticks")
g = sns.catplot(x="minimum_nights", y="price", hue="price_category", data=df_airbnb)
```

```python
sns.set(style="ticks")
g = sns.catplot(x="calculated_host_listings_count", y="price", hue="price_category", data=df_airbnb)
```

Comment:
There are more low priced listings when a host manages a smaller number of properties but they are some present among those hosts with mulitple listings. Listings at middle and high price are almost equally represented among one-listing-hosts and mulitple-listing-hosts.
