## 3. Preparing the data frame to be plotted

I want to look at the regions in Broward County, FL by the average price per night.

```python
# Create a data frame with just the neighbourhoods and price
data_to_plot = df_airbnb[['neighbourhood', 'price']]

data_to_plot.head()
```

```python
# Create a data frame with an average price by using the groupby function

data_to_plot.groupby('neighbourhood', as_index=False)['price'].mean()

avg_price = data_to_plot.groupby('neighbourhood', as_index=False)['price'].mean()

avg_price
```

## 4. Map

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = county_geo, 
    data = avg_price,
    columns = ['neighbourhood', 'price'],
    key_on = 'feature.properties.neighbourhood', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "price").add_to(map)
folium.LayerControl().add_to(map)

map
```
